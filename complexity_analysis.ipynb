```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import os, sys
```

```python
[x.name for x in os.scandir()]
```

```python
for entry in sorted(os.scandir(), key=lambda x: x.name):
    if '.csv' not in entry.name or entry.name[0] != 'D':
        continue
    df = pd.read_csv(entry.name)
    dummy = int(entry.name[1:3])
    input_num = int(entry.name[4:6])
    comp = df['complexity']
    comp_log = np.log(comp)
    plt.figure(figsize=(8, 5))
    sns.kdeplot(comp).set(title=f'Complexity distribution for {input_num} inputs and {dummy} dummies, log scaled')
    plt.show()
```
