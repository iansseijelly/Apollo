```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
comp = df['complexity']
comp_log = np.log(comp)
sns.kdeplot(comp).set(title='Complexity distribution for 7 inputs and 3 dummies, linear scaled')
```

```python
plt.hist(comp_log, bins=40)
print(f'mean of complexity is {np.mean(comp)}')
print(f'median of complexity is {np.median(comp)}')
print(f'std of complexity is {np.std(comp)}')
```
